```python
import pandas as pd

# Input and output file paths
input_file = 'EUCLIDEAN_50.txt'
output_file = 'output.xlsx'

# Read the text file
with open(input_file, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Split each line by spaces
data = [line.strip().split() for line in lines if line.strip()]

# Convert to DataFrame
df = pd.DataFrame(data)

# Save to Excel
df.to_excel(output_file, index=False, header=False)

print(f"✅ Successfully converted '{input_file}' to '{output_file}'")
```

```python
import pandas as pd
import chardet  # to detect file encoding

# Input and output file paths
input_file = 'EUCLIDEAN_50.txt'
output_file = 'output.xlsx'

# Step 1: Detect encoding automatically
with open(input_file, 'rb') as f:
    result = chardet.detect(f.read())
encoding = result['encoding']
print(f"Detected encoding: {encoding}")

# Step 2: Read the text file safely using detected encoding
with open(input_file, 'r', encoding=encoding, errors='ignore') as file:
    lines = file.readlines()

# Step 3: Split each line by spaces
data = [line.strip().split() for line in lines if line.strip()]

# Step 4: Convert to DataFrame
df = pd.DataFrame(data)

# Step 5: Save to Excel
df.to_excel(output_file, index=False, header=False)

print(f"✅ Successfully converted '{input_file}' to '{output_file}'")
```
